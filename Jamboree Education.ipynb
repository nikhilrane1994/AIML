```python
# Statmodels implementation of Linear regression 
import statsmodels.api as sm
# Lets scale the data, standardization 
from sklearn.preprocessing import StandardScaler

#X = df[df.columns.drop('Chance of Admit')]
X=df.drop(['Chance of Admit'],axis=1)
Y = df["Chance of Admit"] 
sc = StandardScaler()
cols = X.columns
X[cols] = sc.fit_transform(X[cols])

X_sm = sm.add_constant(X)  #Statmodels default is without intercept, to add intercept we need to add constant 
sm_model = sm.OLS(Y, X_sm).fit()
print(sm_model)
print(sm_model.summary())

print("coefficient of features",sm_model.params)
```

```python
#Display model coefficients with column names
print("coefficient of features",sm_model.params)
```

```python

print(X_sm.shape)
```

```python
#Multicollinearity check by VIF score (variables are dropped one-by-one till none has VIF>5) (10 Points)
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
X_t = X
vif['Features'] = X_t.columns
vif['VIF'] = [variance_inflation_factor(X_t.values, i) for i in range(X_t.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
#The mean of residuals is nearly zero
```

```python
# Linearity of variables (no pattern in the residual plot) (10 Points)
#The mean of residuals is nearly zero (10 Points)

Y_hat= sm_model.predict(X_sm)

errors = Y_hat - Y 
import seaborn as sns

sns.kdeplot(errors)
plt.xlabel(" Residuals")
plt.title("Histogram of residuals")
```

```python
#Test for Homoscedasticity (10 Points)
sns.scatterplot(Y_hat,errors)
plt.xlabel("predicted Selling price")
plt.ylabel("Residuals")
plt.title("Predicted values vs Residuals")
```
